```python
import pandas as pd
import numpy as np
```

```python
# Import raw data from csv.
df = pd.read_csv('./RawData.csv')
df.head()
```

```python
# Replace weird null values with native numpy NaN.
df.replace(to_replace="n.a.", value=np.nan, inplace=True)
df.replace(to_replace="n.s.", value=np.nan, inplace=True)

df.isna().sum()
```

```python
# Drop unnecessary index column.
df.drop(['Unnamed: 0'], axis=1, inplace=True)
# Drop rows with NaN values.
df.dropna(inplace=True)
df.shape[0]
```

```python
# Rename columns.
ends = ["\nth GBP", "\n"]
for col in df.columns:
    for end in ends:
        if end in col:
            df.rename(columns={col: col.replace(end, "")}, inplace=True)

df.columns
```

```python
# Print dtypes of columns.
df.dtypes
```

```python
# Get rid of commas in numerically formatted columns.
# Try to convert data to numeric, float if there is decimal point, integer otherwise.
for col in df.columns:
    df[col] = df[col].str.replace(',', '')
    try:
        df[col] = pd.to_numeric(df[col], errors='raise')
    except ValueError:
        pass

df.dtypes
```

```python
# Feature Engineering
years = ["2015", "2016", "2017", "2018"]

for col in df.columns:
    if "Turnover" in col:
        df[col] = np.log(df[col])
        df.rename(columns={col: col.replace("Turnover", "Log Sales")}, inplace=True)
    
for year in years:
    # Profibility variable.
    df["Profitability " + year] = df["EBITDA " + year] / df["Total Assets " + year]
    # Non-debt tax shields proxy.
    df["NDTS" + year] = df["Depreciation " + year] / df["Total Assets " + year]
```

```python
df.columns
```

```python
# Calculate column averages for years 2015-2017.
df["Asset Tangability Avg"] = df[["Asset Tangibility2015", "Asset Tangibility2016", "Asset Tangibility2017"]].mean(axis=1)
df["Directors' Fees Avg"] = df[["Directors' Fees 2015", "Directors' Fees 2016", "Directors' Fees 2017"]].mean(axis=1)
df["Directors Remuneration Avg"] = df[["Directors' Remuneration 2015", "Directors' Remuneration 2016", "Directors' Remuneration 2017"]].mean(axis=1)
df["Highest Paid Director Avg"] = df[["Highest Paid Director 2015", "Highest Paid Director 2016", "Highest Paid Director 2017"]].mean(axis=1)
df["Log Sales Avg"] = df[["Log Sales 2015", "Log Sales 2016", "Log Sales 2017"]].mean(axis=1)
df["TobinsQ Avg"] = df[["Market capitalisation / Total assets (Tobin’s Q)2015", "Market capitalisation / Total assets (Tobin’s Q)2016", "Market capitalisation / Total assets (Tobin’s Q)2017"]].mean(axis=1)
df["Profitability Avg"] = df[["Profitability 2015", "Profitability 2016", "Profitability 2017"]].mean(axis=1)
df["NDTS Avg"] = df[["NDTS2015", "NDTS2016", "NDTS2017"]].mean(axis=1)
pd.set_option('display.max_columns', None)
df.head()
```

```python
# Save whole dataframe for analysis.
df.to_csv('./CleanData.csv', index=False)

# Save dataframe for regression.
regression_df = df[["Gearing (%)2018", "Asset Tangability Avg", "Directors' Fees Avg", "Directors Remuneration Avg", "Highest Paid Director Avg", "Log Sales Avg", "TobinsQ Avg", "Profitability Avg", "NDTS Avg"]]

# Drop outliers perhaps?

regression_df.to_csv('./RegressionData.csv', index=False)
```
