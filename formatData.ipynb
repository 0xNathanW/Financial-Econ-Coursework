```python
import pandas as pd
import numpy as np
```

```python
# Import raw data from csv.
df = pd.read_csv('./RawData.csv')
df.head()
```

```python
# Replace weird null values with native numpy NaN.
df.replace(to_replace="n.a.", value=np.nan, inplace=True)
df.replace(to_replace="n.s.", value=np.nan, inplace=True)

df.isna().sum()
```

```python
# Drop unnecessary index column.
df.drop(['Unnamed: 0'], axis=1, inplace=True)
# Drop rows with NaN values.
df.dropna(inplace=True)
df.shape[0]
```

```python
# Rename columns.
ends = ["\nth GBP", "\n"]
for col in df.columns:
    for end in ends:
        if end in col:
            df.rename(columns={col: col.replace(end, "")}, inplace=True)

df.columns
```

```python
# Print dtypes of columns.
df.dtypes
```

```python
# Get rid of commas in numerically formatted columns.
# Try to convert data to numeric, float if there is decimal point, integer otherwise.
for col in df.columns:
    df[col] = df[col].str.replace(',', '')
    try:
        df[col] = pd.to_numeric(df[col], errors='raise')
    except ValueError:
        pass

df.dtypes
```

```python
# Feature Engineering
years = ["2015", "2016", "2017", "2018", "2019"]

for col in df.columns:
    if "Turnover" in col:
        df[col] = np.log(df[col])
        df.rename(columns={col: col.replace("Turnover", "Size")}, inplace=True)
    elif "Gearing (%)" in col:
        df.rename(columns={col: col.replace("Gearing (%)", "Gearing ")}, inplace=True)
    elif "EBITDA margin (%)" in col:
        df.rename(columns={col: col.replace("EBITDA margin (%)", "Profitability ")}, inplace=True)
    elif "Total Assets (th GBP)" in col:
        df.rename(columns={col: col.replace("Total Assets (th GBP)", "Growth Opportunities")}, inplace=True)
    
for year in years:
    # Non-debt tax shields proxy.
    df["NDTS " + year] = df["Depreciation " + year] / df["Total Assets " + year]
    df.drop(["Depreciation " + year], axis=1, inplace=True)
    # Asset Tangability.
    df["Asset Tangability " + year] = df["Fixed Assets " + year] / df["Total Assets " + year]
    df.drop(["Fixed Assets " + year], axis=1, inplace=True)
    df.drop(["Total Assets " + year], axis=1, inplace=True)

df.columns
```

```python
df.head()
```

```python
# make average of all years
df["Gearing"] = df[["Gearing 2018", "Gearing 2017", "Gearing 2016", "Gearing 2015"]].mean(axis=1)
df["Size"] = df[["Size 2018", "Size 2017", "Size 2016", "Size 2015"]].mean(axis=1)
df["Asset Tangability"] = df[["Asset Tangability 2018", "Asset Tangability 2017", "Asset Tangability 2016", "Asset Tangability 2015"]].mean(axis=1)
df["Growth Opportunities"] = df[["Growth Opportunities 2018", "Growth Opportunities 2017", "Growth Opportunities 2016", "Growth Opportunities 2015"]].mean(axis=1)
df["Profitability"] = df[["Profitability 2018", "Profitability 2017", "Profitability 2016", "Profitability 2015"]].mean(axis=1)
df["NDTS"] = df[["NDTS 2018", "NDTS 2017", "NDTS 2016", "NDTS 2015"]].mean(axis=1)
df.head()
```

```python
df.drop(df[(df["BvD Independence Indicator"] == "-") | (df["BvD Independence Indicator"] == "U")].index, axis=0, inplace=True)
df.head()
```

```python
# Save whole dataframe for analysis.
df.to_csv('./CleanData.csv', index=False)

# Save dataframe for regression.
regression_df = df[[
    "Gearing 2019", 
    "BvD Independence Indicator",
    "Asset Tangability", 
    "Size", 
    "Growth Opportunities", 
    "Profitability", 
    "NDTS"
    ]]
# Drop outliers perhaps?
#regression_df = regression_df[regression_df["TobinsQ"] < 15]

regression_df["Ownership_Concentrated"] = regression_df["BvD Independence Indicator"].map({
    "A+": 0, "A-": 0, "A": 0,
    "B+": 0, "B-": 0, "B": 0,
    "C+": 1, "C-": 1, "C": 1,
    "D+": 1, "D-": 1, "D": 1,
})

regression_df.drop(["BvD Independence Indicator"], axis=1, inplace=True)


regression_df.rename(columns={
    'Gearing 2019': 'Gearing',
    "Asset Tangability": "Asset_Tangability",
    "Growth Opportunities": "Growth_Opportunities",
    "NDTS": "Non_Debt_Tax_Shields",
    }, inplace=True)

# replace spaces in column names with underscores.
regression_df.columns = [col.replace(" ", "_") for col in regression_df.columns]
regression_df.to_csv('./RegressionData.csv', index=False)
regression_df.head()
```
